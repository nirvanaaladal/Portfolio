```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
files = os.listdir(os.path.join("Data","Train"))
files
```

```python
for file in files:
    images = os.listdir(os.path.join("Data","Train",file))
    count = 1
    for image in images:
        try:
            img = cv2.imread(os.path.join("Data","Train",file,image))
            img = cv2.resize(img,(128,128))
            cv2.imwrite(os.path.join("Data","Train2",file,image),img)
            print("\r",file,'-',count)
            count += 1
        except:
            pass
```
